```python
import numpy as npimport matplotlib.pyplot as pltdef gerade(x,y):    x=np.array(x)    y=np.array(y)    b=np.sum((x-np.mean(x))*y)/sum((x-np.mean(x))**2)    a=np.mean(y)-b*np.mean(x)    n=len(x)    sy=np.sqrt(np.sum((y-b*x-a)**2)/(n-2))    sa=sy*np.sqrt(1./n+np.mean(x)**2/np.sum((x-np.mean(x))**2))    sb=sy*np.sqrt(1./np.sum((x-np.mean(x))**2))    return (a,b,sa,sb,sy)def pplot(lk,slk,f,sf):    k=np.arange(1,len(lk)+1,1)    plt.figure(figsize=(8, 8))    plt.errorbar(k,lk, yerr=slk,fmt='.')    (a,b,sa,sb,sy)=gerade(k,lk)    #ausgleichsgerade    plt.plot([0,len(lk)+1],[a,a+b*(len(lk)+1)])    plt.figsize(10,10)    plt.xlim([0,len(lk)+1])    plt.ylim([0,max(lk)*1.1])    plt.title(r"Quinckesches Resonanzrohr bei "+str(f)+"Hz",fontsize=16)    lamb=f*b/50    slamb=lamb*np.sqrt((sf/f)**2+(sb/b)**2)/50    plt.xlabel(r"$a="+str(a)+"$   $b="+str(b)+"$\n$S_a: "+str(sa)+"$   $S_b: "+str(sb)+"$   $S_y: "+str(sy)+"$\n$\lambda="+str(lamb)+"$  $S_\lambda="+str(slamb)+"$",fontsize=16)    plt.grid()    plt.ylabel("gemessene Distanz")    plt.savefig(str(f)+"Hz.pdf", orientation='portrait')    plt.show()    return lamb
```
