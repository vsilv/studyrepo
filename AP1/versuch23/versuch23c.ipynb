```python
import numpy as npimport matplotlib.pyplot as pltdef gerade(x,y):    x=np.array(x)    y=np.array(y)    b=np.sum((x-np.mean(x))*y)/sum((x-np.mean(x))**2)    a=np.mean(y)-b*np.mean(x)    n=len(x)    sy=np.sqrt(np.sum((y-b*x-a)**2)/(n-2))    sa=sy*np.sqrt(1./n+np.mean(x)**2/np.sum((x-np.mean(x))**2))    sb=sy*np.sqrt(1./np.sum((x-np.mean(x))**2))    return (a,b,sa,sb,sy)def pplot(lk,slk,t,st,f,sf):    plt.figure(figsize=(8, 8))    plt.errorbar(t,lk, yerr=slk,xerr=st,fmt='.')    (a,b,sa,sb,sy)=gerade(t,lk)    #ausgleichsgerade    plt.plot([0,max(t)*1.1],[a,a+b*max(t)*1.1])    plt.figsize(10,10)    plt.xlim([0,max(t)*1.1])    plt.ylim([0,max(lk)*1.1])    plt.title(u"Ultraschall/ Messung der Laufzeit bei "+str(f)+"Hz",fontsize=16)    plt.xlabel(r"t / ms         $a="+str(a)[0:6]+"$ $cm$   $b="+str(b*20)[0:6]+"$ $m/s$\n$S_a= "+str(sa)[0:6]+"$ $cm$   $S_b= "+str(sb*20)[0:6]+"$ $m/s$   $S_y= "+str(sy)[0:6]+"$ $cm$",fontsize=16)    plt.grid()    plt.ylabel("gemessene Distanz / cm",fontsize=16)    plt.savefig(str(f)+"Hz.pdf", orientation='portrait')    plt.show()
```
